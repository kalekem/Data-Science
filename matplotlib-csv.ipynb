```python
from matplotlib import pyplot as plt
from matplotlib import style
import numpy as np

style.use('ggplot')

x, y = np.loadtxt('test.csv',
                 unpack = True,
                 delimiter =',')
plt.plot(x,y)
plt.title('Awesome data')
plt.xlabel('X axis')
plt.ylabel('Y label')

plt.show()
```
